```python
from bs4 import BeautifulSoup
import requests
from csv import writer

url = "https://www.buyrentkenya.com/property-for-sale/nairobi"
page = requests.get(url)
#print(page)

soup = BeautifulSoup(page.content, 'html.parser')

lists = soup.find_all('div', class_="mb-3 w-full")

with open('housing.csv','w', encoding='utf8', newline='') as f:#r-read only,a-end a previous file,w-overwrite previous versions
    thewriter = writer(f)
    header = ['Location','Price','Desription','Baths']
    thewriter.writerow(header)
    for list in lists:
      location = list.find('p', class_="text-md md:text-sm font-normal text-grey-darker mt-1 md:mt-0").text.replace('\n', '')
      price = list.find('a', class_="no-underline").text.replace('\n', '')
      description = list.find('a', class_="no-underline text-black").text.replace('\n', '')
      baths = list.find('span', class_="text-sm").text.replace('\n', '')
      
      info = [location, price, description, baths]
      thewriter.writerow(info)
```

```python
#Remove the KSh
Price1 = housingKe['Price'].str[4:]
print(Price1)
print(type(Price1))
```

```python
#Replace the column
housingKe['Price'] = Prices['Price']
housingKe
```
